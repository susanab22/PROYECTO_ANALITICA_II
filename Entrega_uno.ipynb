#### Union de las tablas

```python
df = df_employee_survey.merge(df_general, on="EmployeeID")
df = df.merge(df_manager_survey, on="EmployeeID")
df = df.merge(df_time_work, on="EmployeeID")

df.info()
df.head(3)
```

Reemplazar valores nulos con la media

```python
df_3 = df.copy()
```

```python
# Variable: Nivel de satisfaccion del entorno
mean_environment_satisfaction = df_3['EnvironmentSatisfaction'].mean()

### Reemplazar los valores nulos con la media/promedio
df_3['EnvironmentSatisfaction'].fillna(mean_environment_satisfaction, inplace=True)

# Variable: Nivel de satisfaccion laboral
mean_JobSatisfaction = df_3['JobSatisfaction'].mean()

### Reemplazar los valores nulos con la media/promedio
df_3['JobSatisfaction'].fillna(mean_JobSatisfaction, inplace=True)

# Variable: Nivel de conciliación de la vida laboral y familiar
mean_WorkLifeBalance = df_3['WorkLifeBalance'].mean()

### Reemplazar los valores nulos con la media/promedio
df_3['WorkLifeBalance'].fillna(mean_WorkLifeBalance, inplace=True)

# Variable: Número total de empresas en las que ha trabajado el empleado
mean_NumCompaniesWorked = df_3['NumCompaniesWorked'].mean()

### Reemplazar los valores nulos con la media/promedio
df_3['NumCompaniesWorked'].fillna(mean_NumCompaniesWorked, inplace=True)

df_3.info()
df_3.head(3)
```
